```python
def target_params(posterior_path, duration, wf_list, srate = 2**14, t_offset=0.0, refifo = "H1",ifo_list = []):
    #method to calculate fit params, returns ones necessary to write config files for analysis 
    #and saves anything else useful in a json
    #refifo is the ifo used for timing, default H1 unless a specific event has a reason not to
    postsamps_list = []
    final_mass_non_evolved= -1.0
    fref = -1.0
    flow = -1.0
    evt_name = posterior_path.split("PosteriorData/")[1].split(".")[0]
    
    if posterior_path.endswith(".dat"):
        wf_ = "C01:IMRPhenomPv2"
        with open(posterior_path,newline='') as postsamps_dat:
            reader = csv.DictReader(postsamps_dat,delimiter='\t')
            #print("reader len =")
            #print(sum(1 for row in reader))
            for row in reader:
                postsamps_list.append({})
                for k in row.keys():
                    postsamps_list[-1][k]=float(row[k])
        if evt_name == "GW170817":#see https://journals.aps.org/prx/pdf/10.1103/PhysRevX.9.031040 for more info, 
                                  #pg. 30 for flow and 13 for fref
                flow = 23
        elif evt_name == "GW170818":
                flow = 16
        elif evt_name == "GW170608":
                refifo = "L1"#the Hanford analysis started at 30 Hz... using L1 instead as a precaution
                flow=20
        else:
            flow = 20
        fref = 20
    elif posterior_path.endswith(".h5"):
        wave_flag = False
        with h5py.File(posterior_path, 'r') as f:
            print("Possible waveforms for this event include " + str(sorted([q for q in f if "comoving" not in q and "PhenomD" not in q],reverse=True)))
            for i in wf_list:
                for waveform in sorted([q for q in f if "comoving" not in q and "PhenomD" not in q],reverse=True):
                    if wave_flag == False:
                        if i in waveform:
                            try:
                                print("Trying to use " + waveform)
                                wf_ = waveform
                                postsamps_list = np.array(f['{w}/posterior_samples'.format(w=wf_)])
                                if len(f["{w}/config_file".format(w=wf_)])!=0:
                                    list(f["{w}/config_file".format(w=wf_)])[0] #check that waveform has config_file
                                    print( "Making fit config file using postsamples from waveform "+ wf_)
                                else:
                                    print( "Making fit config file using postsamples from waveform "+ wf_)
                                wave_flag = True
                            except Exception:
                                print(waveform + " passed over")
                                pass
            try:
                postsamps_list[0]
            except:
                raise OSError("Posterior samples not in known location")
            if len(f["{w}/config_file".format(w=wf_)])==0:
                fref = 20.0 #Default to 20 Hz when config_file is empty, as per https://dcc.ligo.org/public/0170/P2000318/008/o3b_catalog.pdf, page 56
                flow = [20.0]
                print("No {w}/config_file found for event ".format(w=wf_) + evt_name)
            else:
                try: # first assume fref is a single value
                    fref = float(f.get('{w}/config_file/engine/fref'.format(w=wf_))[0].decode("UTF-8"))
                except Exception:
                    #fref = float(f.get("{w}/config_file/config/reference-frequency".format(w=wf_))[()][0].decode("UTF-8"))
                    fref = float(ast.literal_eval(f.get('{w}/config_file/config/reference-frequency'.format(w=wf_))[0].decode("UTF-8")))
                except Exception: # then assume it is a dict
                    fref = ast.literal_eval(f.get('{w}/config_file/config/reference-frequency'.format(w=wf_))[0].decode("UTF-8"))
                    print(fref)
                    for k,v in fref.items():
                        fref[k]=float(v)
                except:
                    raise TypeError("fref type not float")
                try:
                    flow = ast.literal_eval(f.get('{w}/config_file/lalinference/flow'.format(w=wf_))[0].decode("UTF-8"))
                    for k,v in flow.items():
                        flow[k]=float(v)
                #except Exception:
                #    flow = ast.literal_eval(f.get('{w}/config_file/config/minimum-frequency'.format(w=wf_))[0].decode("UTF-8"))
                #    for k,v in flow.items():
                #        flow[k]=float(v)
                except Exception:
                    flow = ast.literal_eval(f.get('{w}/config_file/config/minimum-frequency'.format(w=wf_))[0].decode("UTF-8").replace("H1","\"H1\"").replace("L1","\"L1\"").replace("V1","\"V1\""))
                    for k,v in flow.items():
                        flow[k]=float(v)
                #except Exception:
                #    flow = [float(f.get('{w}/config_file/lalinference/flow'.format(w=wf_))[()][0].decode("UTF-8"))]
                except:
                    raise TypeError("flow type not dict")
                    
    postsamps_list = pd.DataFrame(postsamps_list)
    
    #get final_mass, to set priors
    try:
        final_mass_non_evolved = np.median(postsamps_list["final_mass_non_evolved"])
    except Exception:
        final_mass_non_evolved = np.median(postsamps_list["final_mass"]) #GW190521 does not have final_mass_non_evolved
    except:
        raise ValueError("can't find key for final mass in postsamps_list")
    #get peak time (t_geo + t0)
    ts_peak_geo = []
    ts_peak_dict = []
    ts_geo_samps = []
    hps = []
    hcs = []
    wf_ls = ls.GetApproximantFromString(wf_.replace("C01:",""))
    if not posterior_path.endswith(".dat"):#Treat GWTC-1 differently since fref is not read from posteriors
        try:
            flow_max=float(max(flow.values()))
        except Exception:
            flow_max = float(max(flow))
        except:
            raise TypeError("flow type not float or dict")
        if type(fref)!=float:
            try:
                fref_max=float(max(fref.values()))
            except Exception:
                fref_max = float(max(fref))
            except:
                raise TypeError("fref type not float or dict")
        else:
            fref_max = fref
    else:
        fref_max = fref
        flow_max = flow
    if "NRSur" in wf_:
        flow_max = -1
    random_indices = rand.sample(postsamps_list.index.to_list(),1000)
    for index in random_indices:
        s = postsamps_list.iloc()[index]
        #if i == 0:
        #    print(s)
        #    print(wf_ls)
        #    print(srate)
        #    print(fref)
        #    print(flow_max)
        t_peak_geo, t_peak_dict, hp, hc = ringdown.complex_strain_peak_time_td(s, wf=wf_ls, dt=1/srate, f_ref=fref_max, f_low=flow_max)
        ts_peak_geo.append(t_peak_geo)
        ts_peak_dict.append(t_peak_dict)
        ts_geo_samps.append(s)
        #ts_geo_samps = pd.concat([ts_geo_samps,s])
        hps.append(hp)
        hcs.append(hc)
        
        #wv= ringdown.Coalescence.from_parameters(time=time_arr,model="NRSur7dq4",f_ref=20.0,f_low=0.0,manual_epoch = False,window=0.0,**samp_dict)
    ts_peak_geo = np.array(ts_peak_geo)
    ts_peak_refifo = np.array([d[refifo] for d in ts_peak_dict])
    tm = np.median(ts_peak_refifo)+t_offset
    im = np.argmin(abs(ts_peak_refifo - tm))
    t_geo = ts_peak_geo[im]
    median_postsample = ts_geo_samps[im]
    t_peak_dict = ts_peak_dict[im]
    hp = hps[im]
    hc = hcs[im]
    
    # get ra, dec, psi
    ra = ts_geo_samps[im]["ra"]
    dec = ts_geo_samps[im]["dec"]
    psi = ts_geo_samps[im]["psi"]
    
    #save median sample antenna patterns
    lal_det = {} #ifo information
    antenna_patterns = {}
    gmst = lal.GreenwichMeanSiderealTime(t_geo)
    for ifo in ifo_list:
        lal_det[ifo] = lal.cached_detector_by_prefix[ifo]
        antenna_patterns[ifo] = lal.ComputeDetAMResponse(lal_det[ifo].response, ra, dec, psi, gmst)

    ts_geo_samps_df = pd.DataFrame()#make a pandas dataframe holding samples
    for i in ts_geo_samps:
        ts_geo_samps_df=pd.concat([ts_geo_samps_df,i],axis=1)
    misc_param_dict={"ts_peak_dict": ts_peak_dict, "fref": fref, "flow": flow, "refifo": refifo, "wf":wf_ls, 
                     "ts_geo_samps_df":ts_geo_samps_df.to_json(), "median_postsample":median_postsample.to_json(), "hp_median":pd.Series(hp.data.data).to_json(), "hc_median":pd.Series(hc.data.data).to_json(), "antenna_patterns":antenna_patterns}

    return {"t_geo":t_geo, "tm_refifo":tm,"ra":ra,"dec":dec,"psi":psi,"M":final_mass_non_evolved,"miscellaneous":misc_param_dict}
```

```python
def save_param_json(misc_param_dict,evt_name,duration,t_offset,srate,mkdir=False):
    json_str = json.dumps(misc_param_dict)
    if mkdir:
        subprocess.run(["mkdir",evts_dir + "/{e}".format(e=evt_name)]) # make directory for output of params info, if dir exists already nothing should happen
    with open(evts_dir+"/{e}/fit_misc_param_dict_{t}_{s}_{L}.json".format(e=evt_name,t='t0_'+str(t_offset),s="sr"+str(srate),L="T_"+str(duration)),"w") as f:
        f.write(json_str)
```

```python
T = 0.1 # analysis choices/other parameters
sr = 4096
t_offset=0.0
modes=[(1, -2, 2, 2, 0), (1, -2, 2, 2, 1)]
flat_A=1
flat_A_ellip = 0
max_hz = 16384
NS_events_list = ["GW170817","GW190425","GW200105","GW200115_042309","GW190814","GW191219_163120","GW200210_092254"] #events that may have neutron star, ignore these
skip_events_list = ["GW190403_051519","GW190426_190642"] #events we cannot use for various reasons
```

```python
for (p,s) in zip(posterior_paths_list,strain_paths_grouped):#loop to make all fit config files for each event
    ifo_strain_dict = {ifo_strain.split(str(strain_dur)+"_s/")[1][2:4]: ifo_strain for ifo_strain in s}
    evt_name = p.split("PosteriorData/")[1].split(".")[0]
    print(p)
    #if evt_name!="GW190727_060333":
    #    continue
    if evt_name in NS_events_list:
        print("Skipping NS event "+ evt_name + "\n")
        continue
    if evt_name in skip_events_list:
        print("Skipping event "+ evt_name + "\n")
        continue
    #if evt_name in os.listdir(evts_dir):
    #    continue
    subprocess.run(["mkdir",evts_dir + "/{e}".format(e=evt_name)]) # make directory for outputs, if dir exists already nothing should happen

    params = target_params(p,srate = sr,duration=T,wf_list = wf_list,ifo_list = ifo_strain_dict.keys())
    save_param_json(params["miscellaneous"],evt_name,srate=sr,duration=T,t_offset=0.0)
    fit = ringdown.Fit(model='mchi', modes=modes)
    fit.load_data(ifo_strain_dict,kind="GWOSC") # TO DO: CHECK IF STRAIN IS NANs
    fit.set_target(t0=params["t_geo"], ra=params["ra"], dec=params["dec"], psi=params["psi"], duration=T)
    fit.condition_data(ds=int(max_hz/sr),digital_filter=True)
    fit.compute_acfs()
    fit.update_prior(A_scale=5e-21, M_min=params["M"]*0.5, M_max=params["M"]*2.0,flat_A=1)
    if flat_A == 1:
        fit.to_config(evts_dir + "/{e}/fit_config_{t}_{s}_{L}_flat_A.ini".format(e=evt_name,t='t0_'+str(t_offset),s="sr"+str(sr),L="T_"+str(T)))
    elif flat_A_ellip == 1:
        fit.to_config(evts_dir + "/{e}/fit_config_{t}_{s}_{L}_flat_A_ellip.ini".format(e=evt_name,t='t0_'+str(t_offset),s="sr"+str(sr),L="T_"+str(T)))
    else:
        fit.to_config(evts_dir + "/{e}/fit_config_{t}_{s}_{L}_gaussprior.ini".format(e=evt_name,t='t0_'+str(t_offset),s="sr"+str(sr),L="T_"+str(T)))
    fit.update_prior(perturb_f=[0,1], perturb_tau=[0,1])
    if flat_A == 1:
        fit.to_config(evts_dir + "/{e}/fittestgr_config_{t}_{s}_{L}_flat_A.ini".format(e=evt_name,t='t0_'+str(t_offset),s="sr"+str(sr),L="T_"+str(T)))
    elif flat_A_ellip == 1:
        fit.to_config(evts_dir + "/{e}/fittestgr_config_{t}_{s}_{L}_flat_A_ellip.ini".format(e=evt_name,t='t0_'+str(t_offset),s="sr"+str(sr),L="T_"+str(T)))
    else:
        fit.to_config(evts_dir + "/{e}/fittestgr_config_{t}_{s}_{L}_gaussprior.ini".format(e=evt_name,t='t0_'+str(t_offset),s="sr"+str(sr),L="T_"+str(T)))
    print(evt_name + " config files successfully written \n")
```
