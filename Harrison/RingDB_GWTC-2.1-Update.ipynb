```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

import ringdown
from ringdb import Database
import lalsimulation as ls
import numpy as np
import pandas as pd

db = Database('./Data')
db.initialize()
```

```python
soup = BeautifulSoup(requests.get("https://zenodo.org/record/6513631").content, "html.parser")

base_url = "https://zenodo.org"
```

```python
a = list(filter(lambda x: x.text == ' Download', soup.find_all('a')))[0]

def get_properties(url):
    try:
        event, cosmo = url.replace('.h5?download=1','').split('-v2-')[1].split('_PEDataRelease_mixed_')
    except (ValueError,IndexError):
        return None
    if cosmo == 'cosmo':
        cosmo = True
    else:
        cosmo = False
    return {'event': event, 'cosmo': cosmo, 'url' : base_url + url}

urls = [get_properties(a.attrs['href']) for a in soup.find_all('a') if (a.text == ' Download') and ('_PEDataRelease_mixed_' in a.attrs['href']) and ('skymap' not in a.attrs['href'])]
```

```python
df_urls = pd.DataFrame(urls)
```
