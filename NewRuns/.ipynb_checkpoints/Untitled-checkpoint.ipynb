```python
import ringdown
from ringdb import Database
import lalsimulation as ls
import numpy as np
import pandas as pd

db = Database('./Data')
db.initialize()
```

```python
event = db.event('GW150914')
strains = event.strain()
posts = event.posteriors()
```

```python
srate = 2**14
fref_max = 20
flow_max = 20
wf_ls = int(posts['waveform_code'].unique()[0])


samps = [x.to_dict() for i,x in posts.sample(801).iterrows()]

for i,x in enumerate(samps):
    t_peak, t_dict, hp, hc = ringdown.complex_strain_peak_time_td(x,
                                                                  wf=wf_ls, dt=1/srate, 
                                                                  f_ref=fref_max, f_low=flow_max)
    samps[i].update({k+'_peak':v for k,v in t_dict.items()})
    
samps = pd.DataFrame(samps)
```

```python
im = (samps['geocent_peak'] - samps['geocent_peak'].median()).abs().argmin()
median_samp = samps.iloc[im]

args = median_samp[['geocent_peak', 'ra','dec','psi']].rename({'geocent_peak':'t0'}).to_dict()
```

```python
t_peak = args['t0']
waveform_model = posts['waveform_name'].unique()[0]
wv = ringdown.Coalescence.from_parameters(time=np.linspace(t_peak-0.05, t_peak+0.05, int(0.1*2**14)),
                                          model=waveform_model,f_ref=20.0,
                                          f_low=20.0,manual_epoch = False,window=0.0, 
                                          **posts.iloc[0].to_dict())
```

```python
fit3 = ringdown.Fit(model='mchi', modes=[(1, -2, 2, 2, 0), (1, -2, 2, 2, 1)])

fit3.add_data(strains['H1'])
fit3.add_data(strains['L1'])

sr = 4096

#fit.set_target(**args, duration=0.1)
#fit2.set_target(1126259462.4083147, ra=1.95, dec=-1.27, psi=0.82, duration=0.1)
fit3.set_target(1126259462.4156113, ra=1.0349596970510788, dec=-1.0763290343573146, psi=1.2780754359053557, duration=0.1)
fit3.condition_data(ds=int(srate/sr),digital_filter=True)
fit3.compute_acfs()

fit3.update_prior(A_scale=5e-21, M_min=median_samp['final_mass']*0.5,
                 M_max=median_samp['final_mass']*2.0,
                 flat_A=1)
```
